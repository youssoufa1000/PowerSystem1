# Run Analysis

## POWER FLOW

**runpp** function is used to run power flow analysis

```python
     pandapower.runpp(net,algorithm='nr')
```

algorithm available

           nr ==== newton raphson
           gs ==== gauss seidel
           fdbx === fast decoupled
           fdxb ==== fast decoupled
           bfsw ==== backward/forward sweep
  
## OPTIMAL FLOW

**runopp** function is used to run optimal power flow

# ACCESS RESULT

After running an analysis the result can be retreive as follow

res_bus =====> retreive buses results

res_line ====> retreive lines results

res_ext_grid ===> retreive external grid results

res_load =====> retreive load result

Basically all elements results can be retreive by just typing **"res_element"**

````python
# run power flow analysis

pp.runpp(net0,algorithm='gs')

print(net0)
````

# Some Useful Function to create user define network

Pandapower is allows to create our own network in an easy manner.Before working with pandapower you must first  import pandapower as pp

```python 
     import pandapower as pp
    @Hence forward it will be assumed the module has been imported
```

## create empty network

This function will allow us to create an empty network then later on will add the elements of the network

```python
     net = pp.create_empty_network()
    @there as some optional parameters such name,frequency, base MVA
```

## create elements

```python
       bus1 = pp.create_bus(net,vn_kv = 20,name ='bus1')  # adding bus1 to network the empty network created
    
    # creating a slack bus add it to the created network
    grid = pp.create_ext_grid(net,bus= b1,vm_pu = 1.0,va_degree=0.0,name='grid')
    
    # creating a load
    load = pp.create_load(net,bus=b3,p_mw=0.1,q_mvar=0.05,name='load')
    
```

### creating a line

a line can be created with either standard type parameters or with user define parameters.

```python

          @create line with standard type
    line1 = pp.create_line(net,from_bus=b1,to_bus=b2,lenght_km=0.1,std_type='NAYY 4x50 SE',
    name='line')
      
    @create line with user define parameters
    line = pp.create_line_from_parameters(net,from_bus=b1,to_bus=b2,length_km=1,
     r_ohm_per_km=0.1,x_ohm_per_km=0.2,c_nf_per_km=0,name='line')
    
```

### creating a transformer 2 windings

like a line a transformer can be build either with user define parameters or standard type parameters.

```python

    @standard type parameters
    
    tid = pp.create_transformer(net, hv_bus=b1, lv_bus=b2, std_type='0.4 MVA 20/0.4 kV', name='trafo')
    
    @user defined parameters
    
    pp.create_transformer_from_parameters(net,hv_bus=b1,lv_bus=b2,sn_mva=40,vn_hv_kv=10,vn_lv_kv=0.4,
           name='trafo')
  ```
  

### creating 3 windings transformer

same methodology as a 2 winding transformer

```python

     @using standard type library
    
    pp.create_transformer3w(net,hv_bus=b2,mv_bus=b1,lv_bus=b3,std_type='63/25/38 MVA 110/10/10 kV',
        name='trafo3w')
    
    @using user define parameters
    
    pp.create_transformer3w_from_parameters(net,hv_bus=b1,mv_bus=b2,lv_bus=b3,vn_hv_kv=100,
            vn_mv_kv=10,vn_lv_kv=10,sn_hv_mva=50,sn_mv_mva=10,sn_lv_mva=10,name='trafo3w')
```
